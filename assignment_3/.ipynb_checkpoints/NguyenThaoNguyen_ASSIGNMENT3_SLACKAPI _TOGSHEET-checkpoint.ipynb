* Tạo thành 1 bảng chứa các thông tin trên và update lên Spreadsheet (Sheet: Assignment#2 Submission)

```python
def request_SlackMessageAPI(channelID):
    """Return the message data of given channelID in Slack Conversation
    
    """
    endpoint = "https://slack.com/api/conversations.history"
    data = {"channel":channelID } ## This is ID of assignment#2 channel
    headers = {"Authorization": "Bearer {}".format(os.environ['SLACK_BEARER_TOKEN'])}
    try:
        response = requests.post(endpoint, data = data, headers = headers)
        if (response.status_code == 200):
            response_json = response.json()
            msg_data = response_json['messages']
            return(msg_data)
        else:
            print("Error in the HTTP request")
            return(None)
    except Exception as e:
        print(e)
        return(None)
    
def extract_MessageData(msg_ls, not_learners_id):
    """Return a desired information in Dictionary form from given msg_data.
    The ouput hold the 'datetime', 'user ID', 'reply_users_count': number of users replying,
    'reply_count': count of view, 'reply_users': list of replying users,
    'latest_reply': datetime of the lastest reply,'github':attachment github link  
    """
    ## Summarize all submitted assignments + reviews cnt
    #not_learners_id = ['U01BE2PR6LU'] # -> Remove MA from the user_id
    github, reply_count, reply_users_count, reply_users, latest_reply = '','','','',''
    mgs_dict = {'datetime':[], 'user':[], 'reply_users_count':[],'reply_count':[],'reply_users':[],
                'latest_reply':[],'github':[]}

    for i in range(len(msg_ls)):
        ts = dt.fromtimestamp(float(msg_ls[i]['ts'])) # -> Convert timestamp Epoch thành dàng dễ đọc
        user = msg_ls[i]['user'] # -> Lấy thông tin người post messages
        if msg_ls[i]['user'] not in not_learners_id:
            if 'attachments' in msg_ls[i].keys():
            #print(msg_ls[i].keys())
                text = msg_ls[i]['text']
                github_link = re.findall('(?:https?://)?(?:www[.])?github[.]com/[\w-]+/?', text) #-> Submission là các message có link github
                #print(msg_ls[i])
                if len(github_link) > 0: github = github_link[0]
                if 'reply_count' in msg_ls[i].keys(): reply_count = msg_ls[i]['reply_count'] #-> Extract số review
                if 'reply_users_count' in msg_ls[i].keys(): reply_users_count = msg_ls[i]['reply_users_count']
                if 'reply_users' in msg_ls[i].keys(): reply_users = msg_ls[i]['reply_users']
                if 'latest_reply' in msg_ls[i].keys(): latest_reply = dt.fromtimestamp(float(msg_ls[i]['latest_reply']))
                #print(ts, user, reply_users_count,reply_count, reply_users, latest_reply, github)

                #-- Ghi dữ liệu vào msg_dic:
                mgs_dict['datetime'].append(ts)
                mgs_dict['user'].append(user)
                mgs_dict['reply_users_count'].append(reply_users_count)
                mgs_dict['reply_count'].append(reply_count)
                mgs_dict['reply_users'].append(reply_users)
                mgs_dict['latest_reply'].append(latest_reply)
                mgs_dict['github'].append(github)  
    return(mgs_dict)
 
def authorizeGSpread(authorizeJsonfile):
    """Return gsrpead authorized by json file
    """
            ## Authorize bằng JSON
    scope = ['https://spreadsheets.google.com/feeds',
             'https://www.googleapis.com/auth/drive']
    credentials = ServiceAccountCredentials.from_json_keyfile_name(
        'diesel-post-314008-065a1771ca30.json', scope)
    gc = gspread.authorize(credentials)
    print("DONE!")
    return(gc)

def writeGspread(gc,df,spreadsheet_key):
    """Write data with dataframe form into given google spreadsheet 
    """
    # ACCES GOOGLE SHEET
    sheet_index_no = 0
    #spreadsheet_key = '1ESgL9IIOUo82tNNm4ZQqpD_meJu-CXKHl4fo4lFfCcs' # input SPREADSHEET_KEY HERE
    sh = gc.open_by_key(spreadsheet_key)
    worksheet = sh.get_worksheet(sheet_index_no) #-> 0 - first sheet, 1 - second sheet etc. 

    # APPEND DATA TO SHEET
    set_with_dataframe(worksheet, mgs_df) #-> Upload user_df vào Sheet đầu tiên trong Spreadsheet
```

```python
#channelID = 'C021FSDN7LJ' ## This is ID of assignment#2 channel
channelID = 'C0227A51SAY' ## This is ID of assignment#3 channel
not_learners_id = ['U01BE2PR6LU']
authorizeJsonfile = 'diesel-post-314008-065a1771ca30.json'
spreadsheet_key = '1ESgL9IIOUo82tNNm4ZQqpD_meJu-CXKHl4fo4lFfCcs'

msg_dict = {}
mgs_ls = request_SlackMessageAPI(channelID)

mgs_dict = extract_MessageData(mgs_ls, not_learners_id)

mgs_df = pd.DataFrame(mgs_dict)

gc = authorizeGSpread(authorizeJsonfile)

writeGspread(gc,mgs_df,spreadsheet_key)
```
